```python
# Install ipywidgets if not already installed
!pip install ipywidgets --quiet

from ipywidgets import interact, Dropdown, FloatText
from IPython.display import display

# Current XE.com exchange rates (NGN as base)
exchange_rates = {
    "NGN": 1,
    "USD": 0.000742,
    "EUR": 0.000630,
    "GBP": 0.000549,
    "JPY": 0.1156
}

currencies = list(exchange_rates.keys())

def convert_currency(amount, from_currency, to_currency):
    # Convert amount
    amount_in_ngn = amount / exchange_rates[from_currency]
    converted_amount = amount_in_ngn * exchange_rates[to_currency]
    print(f"{amount} {from_currency} = {converted_amount:.2f} {to_currency}")

# Create interactive widgets
interact(
    convert_currency,
    amount=FloatText(value=1.0, description="Amount:", step=0.01),
    from_currency=Dropdown(options=currencies, value="NGN", description="From:"),
    to_currency=Dropdown(options=currencies, value="USD", description="To:")
)
```
